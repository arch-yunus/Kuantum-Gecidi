# 🌌 Bölüm 2: Dolanıklık (Entanglement) Oluşturma

Bu dersimizde, Einstein'ın "uzaktan ürkütücü etkileşim" (spooky action at a distance) olarak adlandırdığı **Dolanıklık** kavramını inceleyeceğiz.

## 1. Dolanıklık Nedir?

İki parçacık dolanık hale geldiğinde, aralarındaki mesafe ne kadar olursa olsun, birinin durumu ölçüldüğünde diğerinin durumu anında belirlenir.

### Bell Durumu (Bell State)
En temel dolanık durum olan $|\Phi^+\rangle$ durumunu oluşturmak için şu kapıları kullanırız:
1. **H (Hadamard):** İlk qubiti süperpozisyona sokar.
2. **CNOT (Controlled-NOT):** İlk qubiti kontrol, ikinciyi hedef olarak kullanarak onları birbirine bağlar.

```python
from qiskit import QuantumCircuit, transpile
from qiskit_aer import Aer
from qiskit.visualization import plot_histogram

# 2 qubitlik devre
qc = QuantumCircuit(2)

# 1. Hadamard kapısı
qc.h(0)

# 2. CNOT kapısı
qc.cx(0, 1)

qc.draw(output='mpl')
```

## 2. Ölçüm ve Sonuçlar

Dolanıklığı kanıtlamak için her iki qubiti de ölçelim. Eğer dolanıklarsa, her zaman aynı sonucu vermelerini bekleriz (ya her ikisi de 0, ya her ikisi de 1).

```python
qc.measure_all()

# Simülasyon
backend = Aer.get_backend('qasm_simulator')
job = backend.run(transpile(qc, backend), shots=1024)
result = job.result()
counts = result.get_counts()

plot_histogram(counts)
```

## Sonuç
Histogramda sadece `00` ve `11` durumlarını görüyorsanız, tebrikler! İki qubiti başarıyla birbirine bağladınız. Arada hiç mesafe yokmuş gibi davranacaklar.
