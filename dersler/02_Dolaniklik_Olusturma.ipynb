# 🔗 Bölüm 2: Dolanıklık (Entanglement)

**Dolanıklık**, iki parçacığın öyle bir şekilde birbirine bağlanmasıdır ki, birinin durumu ölçüldüğünde diğerinin durumu (aradaki mesafe ne olursa olsun) anında belirlenir.

### Bell Durumları
En basit dolanık sistemlere **Bell Durumları** denir. Bunlardan en bilineni $\Phi^+$ durumudur:
$$|\Phi^+\rangle = \frac{1}{\sqrt{2}}(|00\rangle + |11\rangle)$$

Bu durumda:
- Eğer ilk qubit 0 ölçülürse, ikinci qubit de kesinlikle 0 olur.
- Eğer ilk qubit 1 ölçülürse, ikinci qubit de kesinlikle 1 olur.

```python
from qiskit import QuantumCircuit

# İki qubitlik devre oluştur
qc = QuantumCircuit(2)

# 1. Qubiti süperpozisyona sok
qc.h(0)

# 2. CNOT (Controlled-NOT) kapısı ile dolanıklık oluştur
qc.cx(0, 1)

print("Bell durumu oluşturuldu.")
qc.draw('mpl')
```
