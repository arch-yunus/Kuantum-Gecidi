# 🔗 Bölüm 2: Dolanıklık (Entanglement) ve Bell Durumları

**Dolanıklık**, Einstein'ın "uzaktan ürkütücü etkileşim" (spooky action at a distance) olarak adlandırdığı, iki parçacığın aralarındaki mesafe ne olursa olsun birbirlerine kopmaz bir bağ ile bağlanmasıdır.

### 1. Bell Durumu Hazırlama
Dolanıklık oluşturmak için en yaygın yöntem **CNOT** (Controlled-NOT) kapısını kullanmaktır. Adımlar şöyledir:
1. Birinci qubit süperpozisyona sokulur ($H$ kapısı).
2. İkinci qubit, birinci qubite bağlanır ($CNOT$ kapısı).

Sonuçta elde edilen en meşhur durum $\Phi^+$ durumudur:
$$|\Phi^+\rangle = \frac{1}{\sqrt{2}}(|00\rangle + |11\rangle)$$

Bu durum şu anlama gelir: Eğer birinci qubiti 0 ölçerseniz, ikincinin 0 olduğu kesindir. Eğer 1 ölçerseniz, ikincinin 1 olduğu kesindir. Arada hiçbir olasılık kalmaz!

```python
from qiskit import QuantumCircuit
from qiskit.visualization import plot_histogram
from qiskit_aer import AerSimulator

# 1. İki qubitlik devre
qc = QuantumCircuit(2)

# 2. Dolanıklık Oluşturma
qc.h(0)
qc.cx(0, 1) # Control: q0, Target: q1

qc.measure_all()

print("Bell devresi şeması:")
display(qc.draw('mpl'))

# 3. Simülasyon
backend = AerSimulator()
counts = backend.run(qc).result().get_counts()
plot_histogram(counts, title="Dolanık Qubitlerin Ölçüm Sonuçları")
```
