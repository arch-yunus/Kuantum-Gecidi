### 1. Temel Gürültü Türleri
- **Dekompozisyon (Decoherence):** Qubitin çevresiyle etkileşime girip kuantum bilgisi kaybetmesi.
  - **T1 (Thermal Relaxation):** Qubitin |1> durumundan |0> durumuna (temel hal) düşmesi.
  - **T2 (Dephasing):** Qubitin süperpozisyon (faz) bilgisini kaybetmesi.
- **Depolarizing Error:** Her işlemde qubitin durumunun rastgele bir yöne kayması.
- **Gate Errors:** Kapı işlemlerinin (H, CNOT vb.) %100 hassasiyetle çalışmaması.
