# ⚛️ Bölüm 1: Qubit Nedir?

Kuantum bilgisayarlarının en temel birimi olan **Qubit** (Kuantum Bit), geleneksel bitlerden (0 veya 1) çok daha fazlasını sunar.

### 1. Süperpozisyon (Superposition)
Bir klasik bit her zaman ya 0 ya da 1 durumundadır. Ancak bir qubit, aynı anda her iki durumun bir kombinasyonu (süperpozisyon) halinde bulunabilir.

### 2. Matematiksel Temsil
Bir qubitin durumu $|\psi\rangle$ (psi) şöyle gösterilir:
$$|\psi\rangle = \alpha|0\rangle + \beta|1\rangle$$
Burada $\alpha$ ve $\beta$ karmaşık sayılardır ve $|\alpha|^2 + |\beta|^2 = 1$ olmalıdır.

## Hadi Kodlayalım!

```python
from qiskit import QuantumCircuit
from qiskit.visualization import plot_bloch_multivector
import numpy as np

# Bir qubitlik devre oluştur
qc = QuantumCircuit(1)

# Qubiti süperpozisyona sok (Hadamard kapısı)
qc.h(0)

print("Hadamard kapısı uygulandı.")
qc.draw('mpl')
```
