# 🌌 Bölüm 1: Qubit Nedir?

Kuantum dünyasına hoş geldiniz! Bu dersimizde, kuantum bilgisayarların temel yapı taşı olan **Qubit** (Kuantum Bit) kavramını inceleyeceğiz.

## 1. Klasik Bit vs. Kuantum Bit

Klasik bir bilgisayarda bilgi `0` veya `1` durumunda bulunur. Kuantum dünyasında ise bir qubit, aynı anda hem `0` hem de `1` durumunun bir kombinasyonunda bulunabilir. Buna **Süperpozisyon** denir.

```python
from qiskit import QuantumCircuit
from qiskit.visualization import plot_bloch_multivector
from qiskit.quantum_info import Statevector

# Bir qubitlik boş bir devre oluşturalım
qc = QuantumCircuit(1)

# Başlangıç durumu (0 durumu)
state = Statevector.from_instruction(qc)
plot_bloch_multivector(state)
```

## 2. Süperpozisyon Oluşturma

Bir qubiti süperpozisyon haline sokmak için **Hadamard (H)** kapısını kullanırız. Bu kapı, qubiti %50 ihtimalle 0, %50 ihtimalle 1 olacağı bir duruma getirir.

```python
qc.h(0) # 0. qubite Hadamard kapısı uygula
state = Statevector.from_instruction(qc)
plot_bloch_multivector(state)
```

## Sonuç
Gördüğünüz gibi, ok artık tam tepede (0) veya tam aşağıda (1) değil; ekvator çizgisi üzerinde bir yerdedir. Bu, qubitin artık süperpozisyon halinde olduğunu gösterir!
