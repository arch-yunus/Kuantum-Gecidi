# ⚛️ Bölüm 1: Qubit ve Kuantum Dünyasına Giriş

Kuantum bilgisayarlarının en temel birimi olan **Qubit** (Kuantum Bit), klasik bilgisayarlardaki bitlerden (0 veya 1) çok daha zengin bir veri temsil kapasitesine sahiptir.

### 1. Süperpozisyon (Superposition)
Klasik bir bit her zaman ya 0 ya da 1 durumundadır (bir anahtarın açık veya kapalı olması gibi). Ancak bir qubit, aynı anda her iki durumun bir lineer kombinasyonu (süperpozisyon) halinde bulunabilir.

### 2. Matematiksel Temsil ve Bloch Küresi
Bir qubitin durumu $|\psi\rangle$ (Dirac gösterimi ile) şöyle ifade edilir:
$$|\psi\rangle = \alpha|0\rangle + \beta|1\rangle$$
Burada $\alpha$ ve $\beta$ karmaşık genliklerdir ve $|\alpha|^2 + |\beta|^2 = 1$ (olasılık korunumu) şartını sağlarlar.

Bu durumu görselleştirmek için **Bloch Küresi** kullanılır. Kürenin kuzey kutbu $|0\rangle$, güney kutbu $|1\rangle$ durumunu temsil ederken, ekvator üzerindeki noktalar süperpozisyon halleridir.

---

```python
from qiskit import QuantumCircuit
from qiskit.visualization import plot_bloch_multivector
from qiskit.quantum_info import Statevector
import matplotlib.pyplot as plt

# 1. Bir qubitlik devre oluştur
qc = QuantumCircuit(1)

# 2. Qubiti süperpozisyona sok (Hadamard kapısı)
# Hadamard (H) kapısı |0> durumunu (|0> + |1>) / sqrt(2) durumuna getirir.
qc.h(0)

print("Hadamard kapısı uygulandı. Devre şeması:")
display(qc.draw('mpl'))

# Durumu görselleştir
state = Statevector.from_instruction(qc)
plot_bloch_multivector(state)
```

### 3. Temel Kuantum Kapıları

Qubitler üzerinde işlem yapmak için kuantum kapıları kullanılır:

1. **Pauli-X (NOT):** Qubiti X ekseni etrafında 180 derece döndürür ($|0\rangle \rightarrow |1\rangle$).
2. **Pauli-Z:** Faz değişimi sağlar ($|1\rangle$ durumunun işaretini değiştirir).
3. **Hadamard (H):** Klasik dünyadan kuantum dünyasına geçiş kapısıdır; süperpozisyon oluşturur.

```python
# X Kapısı Örneği
qc_x = QuantumCircuit(1)
qc_x.x(0)
state_x = Statevector.from_instruction(qc_x)
plot_bloch_multivector(state_x, title="X Kapısı Sonrası Durum (|1>)")
```
