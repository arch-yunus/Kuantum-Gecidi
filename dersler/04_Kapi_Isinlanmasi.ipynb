# 🧪 Bölüm 4: Kuantum Kapı Işınlanması (Gate Teleportation)

Kuantum ışınlanma sadece veri aktarmak için değil, aynı zamanda **hesaplama** yapmak için de kullanılabilir. Bu yönteme "Kapı Işınlanması" (Gate Teleportation) denir.

### 1. Neden Kapı Işınlanması?
- **Hata Toleransı (Fault-Tolerance):** Fragile (hassas) qubitler üzerinde doğrudan kapı uygulamak yerine, kapı işlemini dolanık bir kaynak üzerine "önceden" yükleyip, durumu bu kanal üzerinden ışınlayarak asıl veriyi koruyabiliriz.
- **Evrensel Hesaplama:** T-gate gibi bazı zor kapılar, "Magic State Distillation" yöntemiyle hazırlanıp kapı ışınlanmasıyla devreye dahil edilir.

### 2. Nasıl Çalışır?
Standart ışınlanma devresinde paylaşılan kaynak bir Bell durumudur: $|\Phi^+\rangle$. 
Kapı ışınlanmasında ise kaynağa istediğimiz $U$ kapısını uygularız: $(I \otimes U)|\Phi^+\rangle$. 

Alice durumunu ışınladığında, durum Bob'un elinde $U|\psi\rangle$ olarak belirir.

```python
import sys
import os
sys.path.append(os.path.abspath('../'))

from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_bloch_multivector

# 3 Qubitlik Devre: q0 (Target), q1 (Alice), q2 (Bob)
qc = QuantumCircuit(3, 2)

# Adım 1: Kaynağa T-Gate enjekte et
qc.h(1)
qc.cx(1, 2)
qc.t(2) # Kapı işlemi kanal üzerinde hazırlanıyor
qc.barrier()

# Adım 2: Alice'in Hedef Durumu (|1> yapalım)
qc.x(0)
qc.barrier()

# Adım 3: Işınlanma
qc.cx(0, 1)
qc.h(0)
qc.measure(0, 0)
qc.measure(1, 1)
qc.barrier()

# Adım 4: Bob'un Düzeltmesi
with qc.if_test((qc.clbits[0], 1)):
    qc.z(2)
with qc.if_test((qc.clbits[1], 1)):
    qc.x(2)

print("Kapı Işınlanması Devresi:")
display(qc.draw('mpl'))
```
