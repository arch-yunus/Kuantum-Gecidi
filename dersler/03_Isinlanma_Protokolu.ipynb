# 🌌 Bölüm 3: Kuantum Işınlanma Protokolü

Kuantum ışınlanma, bir qubitin üzerindeki *bilginin* fiziksel bir parçacık taşıma olmaksızın uzak bir noktaya aktarılmasıdır. Bu işlem ışık hızından hızlı veri transferi sağlamaz (hala klasik bir sinyale ihtiyaç duyulur), ancak kuantum ağlarının temelini oluşturur.

### Protokolün Bileşenleri
1. **Alice:** Işınlanacak durumun sahibi.
2. **Bob:** Durumu teslim alacak kişi.
3. **Kanal:** Önceden paylaşılmış bir EPR çifti (dolanık qubitler).

### Adım Adım Akış
1. **Dolanıklık:** Bir EPR çifti oluşturulur ve qubitlerden biri Alice'e, diğeri Bob'a gönderilir.
2. **Alice'in İşlemleri:** Alice, ışınlamak istediği qubit ($|\psi\rangle$) ile kendi dolanık qubitini birleştirir.
3. **Ölçüm ve Klasik Haberleşme:** Alice kendi qubitlerini ölçer. Ölçüm sonucu (00, 01, 10 veya 11) klasik yolla Bob'a iletilir.
4. **Bob'un Düzeltmesi:** Bob aldığı 2-bitlik veriye göre kendi qubitine $X$ veya $Z$ kapıları uygular.

```python
import sys
import os
# Proje kök dizinini ekle (kaynak klasörüne ulaşmak için)
sys.path.append(os.path.abspath('../'))

from kaynak.devre_tasarimci import DevreTasarimci
from kaynak.analiz_araclari import AnalizAraclari
from qiskit import transpile
from qiskit_aer import AerSimulator
from qiskit.quantum_info import Statevector

tasarimci = DevreTasarimci()
qc, qr, crz, crx = tasarimci.isinlanma_devresi_hazirla()

# 1. Işınlanacak durum (Örnek: Belirli bir yön)
theta = 2.3  # rad
phi = 0.5    # rad
tasarimci.durum_hazirla(qc, qr[0], theta, phi)
print(f"Hazırlanan Girdi Durumu: theta={theta}, phi={phi}")

# 2. Protokol Uygulaması
tasarimci.dolaniklik_olustur(qc, qr[1], qr[2])
tasarimci.alice_islemleri(qc, qr[0], qr[1])

qc.measure(qr[0], crz)
qc.measure(qr[1], crx)

tasarimci.bob_duzeltmeleri(qc, qr[2], crz, crx)

print("Işınlanma devresi tamamlandı.")
display(qc.draw('mpl'))
```

### Bob'un Düzeltme Tablosu

Alice'in ölçüm sonuçlarına göre Bob'un uyguladığı işlemler:

| Alice'in Sonucu (crz crx) | Bob'un İşlemi |
| :--- | :--- |
| 00 | Hiçbir şey yapma (ID) |
| 01 | X Kapısı Uygula |
| 10 | Z Kapısı Uygula |
| 11 | Z ve sonra X Kapısı Uygula |
