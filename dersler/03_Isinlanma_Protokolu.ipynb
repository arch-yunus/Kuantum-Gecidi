# 🌌 Bölüm 3: Kuantum Işınlanma Protokolü

Geldik en heyecan verici kısma! Alice'in elindeki bir qubiti Bob'a nasıl gönderebileceğini (ışınlayabileceğini) göreceğiz.

### Protokol Adımları
1.  **Dolanıklık:** Alice ve Bob bir Bell çifti paylaşır.
2.  **Alice'in Hazırlığı:** Alice ışınlamak istediği $|\psi\rangle$ durumunu hazırlar.
3.  **Alice'in Ölçümü:** Alice kendi qubitleri üzerinde Bell ölçümü yapar.
4.  **Haberleşme:** Alice ölçüm sonuçlarını Bob'a klasik yolla (telefon, internet) söyler.
5.  **Bob'un Düzeltmesi:** Bob bu sonuçlara göre kendi qubitine kapılar uygular ve orijinal $|\psi\rangle$ durumunu elde eder.

```python
import sys
import os
# Kaynak klasörü ekle
sys.path.append(os.path.abspath('../'))

from kaynak.devre_tasarimci import DevreTasarimci
from qiskit import transpile
from qiskit_aer import Aer

# Tasarımcıyı kullanarak devreyi kur
tasarimci = DevreTasarimci()
qc, qr, crz, crx = tasarimci.isinlanma_devresi_hazirla()

# Işınlanacak durum (X kapısı ile |1> yapalım)
qc.x(qr[0])

# Protokolü uygula
tasarimci.dolaniklik_olustur(qc, qr[1], qr[2])
tasarimci.alice_islemleri(qc, qr[0], qr[1])
qc.measure(qr[0], crz)
qc.measure(qr[1], crx)
tasarimci.bob_duzeltmeleri(qc, qr[2], crz, crx)

print("Işınlanma devresi hazır.")
qc.draw('mpl')
```
