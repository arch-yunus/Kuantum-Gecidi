# 🌌 Bölüm 3: Kuantum Işınlanma Protokolü

Bu dersimizde, önceki bölümlerde öğrendiğimiz Süperpozisyon ve Dolanıklık kavramlarını birleştirerek tam bir **Kuantum Işınlanma** devresi kuracağız.

## Protokolün Adımları

1. **Hazırlık:** Alice, ışınlamak istediği meçhul $|\psi\rangle$ durumunu hazırlar.
2. **Kanal:** Alice ve Bob bir EPR çifti paylaşır. (Alice'de $q_1$, Bob'da $q_2$)
3. **Alice'in Ölçümü (Bell Ölçümü):** Alice kendi qubitleri üzerinde işlemler yapar ve onları ölçer.
4. **Klasik Bilgi Transferi:** Alice ölçüm sonuçlarını (2 klasik bit) Bob'a telefon veya internet gibi klasik yollarla gönderir.
5. **Bob'un Düzeltmesi:** Bob, aldığı sonuçlara göre kendi qubitine $X$ ve/veveya $Z$ kapısı uygular. Sonuçta $|\psi\rangle$ durumu Bob'a geçmiş olur!

```python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister, transpile
from qiskit_aer import Aer
import numpy as np

# 3 Qubit ve 2 Klasik Bit
qr = QuantumRegister(3, name="q")
crz = ClassicalRegister(1, name="crz")
crx = ClassicalRegister(1, name="crx")
qc = QuantumCircuit(qr, crz, crx)

## 1. ADIM: Işınlanacak Durumu Hazırla
qc.ry(np.pi/4, 0) # Rastgele bir durum
qc.barrier()

## 2. ADIM: Dolanıklık Kanalı
qc.h(1)
qc.cx(1, 2)
qc.barrier()

## 3. ADIM: Alice'in İşlemleri
qc.cx(0, 1)
qc.h(0)
qc.barrier()

## 4. ADIM: Alice'in Ölçümü
qc.measure(0, 0)
qc.measure(1, 1)
qc.barrier()

## 5. ADIM: Bob'un Düzeltmesi
with qc.if_test((crx, 1)):
    qc.x(2)
with qc.if_test((crz, 1)):
    qc.z(2)

qc.draw(output='mpl')
```

## Sonuç
Bu devre başarıyla çalıştırıldığında, $q_0$ üzerindeki durum artık Bob'un elindeki $q_2$ qubitine aktarılmış olacaktır. Orijinal $|\psi\rangle$ ise ölçüm yapıldığı için Alice'in elinde artık mevcut değildir (No-Cloning Teoremi).
